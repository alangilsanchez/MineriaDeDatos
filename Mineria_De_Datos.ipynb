#### Conceptos

**Minería de datos**: Data Mining es el proceso técnico y automático de explorar grandes conjuntos de datos para encontrar patrones, anomalías o correlaciones que no son evidentes a simple vista.

Su objetivo principal es transformar datos brutos en información valiosa que permita predecir comportamientos o tomar decisiones estratégicas fundamentadas.


**Ciencia de datos**: Es un campo interdisciplinario que combina matemáticas, estadística, programación avanzada e inteligencia artificial para extraer conocimientos de grandes volúmenes de datos.

A diferencia de la minería de datos (que se enfoca en encontrar patrones específicos), la ciencia de datos abarca todo el ciclo de vida del dato: desde su recolección y limpieza hasta la creación de modelos predictivos que responden a preguntas complejas de negocio.

#####NumPy Array Attributes

```python
import numpy as np
np.random.seed(0)  # seed for reproducibility
                      #numeros aleatorios entre 0 y 10 (Sin considerar 1ñ 10, python no considera el limite superior)
x1 = np.random.randint(10, size=6)  # One-dimensional array
x2 = np.random.randint(10, size=(3, 4))  # Two-dimensional array ,3 renglones 4 columnas
x3 = np.random.randint(10, size=(3, 4, 5))  # Three-dimensional array
```

```python
print("x3 ndim: ", x3.ndim) #La propiedad ndim nos dice el numero de dimensiones
print("x3 shape:", x3.shape) #
print("x3 size: ", x3.size)
```

```python
print(x2.dtype) # Dtype nos dice el tipo de dato, entero de 64 bits en 8 bloque de bytes
```

```python
#itemsize nos regresa el tamaño en site
print("itemsize:", x3.itemsize, "bytes")
print("nbytes:", x3.nbytes, "bytes") #nbytes es el tamaño de cada elemento por el tamaño
```

###Estadistica Basica


####Medidas Estadísticas Básicas

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

####Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

####Mediana

```python
mediana = np.median(data)
```

####Varianza y Desviación Estándar

```python
varianza = np.var(data)
desviacion = np.std(data) #std = standar
# Desviacion raiz cuadrarda de la varianza
```

####Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000) #media 0, desviacion estandar 1, genera 1000 datos
print(np.mean(muestra), np.std(muestra))
```

####Correlación

```python
corr = np.corrcoef([1,2,3], [1,5,7])
corr
```
