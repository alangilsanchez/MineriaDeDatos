#### Conceptos

**Minería de datos**: Data Mining es el proceso técnico y automático de explorar grandes conjuntos de datos para encontrar patrones, anomalías o correlaciones que no son evidentes a simple vista.

Su objetivo principal es transformar datos brutos en información valiosa que permita predecir comportamientos o tomar decisiones estratégicas fundamentadas.


**Ciencia de datos**: Es un campo interdisciplinario que combina matemáticas, estadística, programación avanzada e inteligencia artificial para extraer conocimientos de grandes volúmenes de datos.

A diferencia de la minería de datos (que se enfoca en encontrar patrones específicos), la ciencia de datos abarca todo el ciclo de vida del dato: desde su recolección y limpieza hasta la creación de modelos predictivos que responden a preguntas complejas de negocio.

```python
# Descriptivos de edad (NumPy)
#Diccionario de python
summary_age = {
    'n': int(edad.size), #total de elementos en el arreglo
    'min': float(np.min(edad)), #el menor elemento del arreglo
    'p25': float(np.percentile(edad, 25)), #percentil genera 100 intervalos y trae el porcentaje de los datos
    'mediana': float(np.median(edad)), #Ordena los datos de menor a mayor y toma el numero de en medio
    'media': float(np.mean(edad)), #promedio, #la media es sensible a datos aberrantes
    'p75': float(np.percentile(edad, 75)),
    'max': float(np.max(edad)), #El maximo elemento
    'rango(ptp)': float(np.ptp(edad)), #peak to peak, la diferencia del mayor menos el menor
    'var_muestral': float(np.var(edad, ddof=1)),#delta degrees of freedom, trae la varianza
    'std_muestral': float(np.std(edad, ddof=1)), #desviacion estandar es una medida de dispersion
    'cv': float(np.std(edad, ddof=1)/np.mean(edad)), #coeficiente de varianza, divide la desviacion estandar entre la media
}
```

```python
# Frecuencias (NumPy) Cantidad de elementos
#Diccionario de las frecuencias de python
#Contamos los que no son 0, es decir los que son true
freq_python = {'Si': int(np.count_nonzero(prog_py)), 'No': int(edad.size - np.count_nonzero(prog_py))}

freq_colab  = {'Si': int(np.count_nonzero(usa_colab)), 'No': int(edad.size - np.count_nonzero(usa_colab))}
freq_gh     = {'Si': int(np.count_nonzero(cuenta_gh)), 'No': int(edad.size - np.count_nonzero(cuenta_gh))}
```

```python
# Tablas cruzadas (NumPy)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int) #numpy.zeros genera una matriz de 2x2 rellenas de 0, y definimos el tipo
    t[1,1] = np.count_nonzero(a & b) #En la posicion [1,1] le asignamos la cuenta de los que no sean sero de a y b
    t[1,0] = np.count_nonzero(a & ~b)
    t[0,1] = np.count_nonzero(~a & b)
    t[0,0] = np.count_nonzero(~a & ~b)
    return t


tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```
